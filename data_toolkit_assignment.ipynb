```python
#3. Explain the following termsg
# The difference in nparray, npasarray and npasanyarray
# The difference between Deep copy and shallow copy
```

```python
#ans:the difference in np.array(), np.asarray(), and np.asanyarray() is listed below:
#1 np.array(): always creates a new array, even if the input is already a numpy array.
#example:
x=[0,1]
arr1=np.array(x)
print(x)
#2.np.asarray():converts input to a numpy but does not copy if the input is already an array.
#example
arr2=np.asarray(arr1)
print(arr2)
#3.np.asanyarray():similar to np.asarray() but it preserves subclasses
#example:
mat= np.matrix([[1, 2], [3, 4]])
arr3 = np.asanyarray(mat)  # Keeps as matrix
print(arr3)
```

```python
# The difference between Deep copy and shallow copy
#ans: the  difference between shallow copy and deep copy is listed below:
#1 shallow copy:copies only the outer object, so changes in nested elements affect both copies.
import copy
arr=[1,2,3,[1,2]]
shallow_copy=copy.copy(arr)
shallow_copy[3][0]=10
print(arr)

#Deep copy: copies the objects and all nested elements so changes don't effect the original.
deep_copy=copy.deepcopy(arr)
deep_copy[3][1]=20
print(arr)
```

```python
#17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant
```

```python
np.random.seed(0)
x=np.random.randn(100)
y=np.random.randn(100)

data=pd.DataFrame({
    'X':x,
    'Y':y
})
def get_quardant(col):
  if col["X"]>=0 and col['Y']>=0:
    return 'quardant-1'
  elif col["X"]<0 and col['Y']>=0:
    return 'quardant-2'
  elif col["X"]<0 and col['Y']<0:
    return 'quardant-3'
  else:
    return 'quardant-4'
data['Quardent']=data.apply(get_quardant,axis=1)

sns.scatterplot(x="X",y="Y",data=data,hue='Quardent')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quardant')

plt.legend(title='Quardant')

plt.show()



```

```python
#19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'?

from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, HoverTool
import pandas as pd
import numpy as np

# Generate random categorical data
np.random.seed(42)  # For reproducibility
categories = [f'Category {i}' for i in range(1, 6)]
values = np.random.randint(1, 100, len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create the figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color='blue')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
output_file("random_categorical_bar_chart.html")
show(p)
```
